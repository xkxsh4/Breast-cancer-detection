```python
#feature scaling 
#split our data set into independent and dependent data sets
X = df.iloc[:,2:31].values  #independent 
Y = df.iloc[:,1].values     #dependent


```

```python
#80:20 ratio 
from sklearn.model_selection import train_test_split
X_train, X_test , Y_train, Y_test = train_test_split(X,Y, test_size=0.20, random_state=0 )
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
def model (X_test, Y_train):
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)
    
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion= 'entropy', random_state=0)
    tree.fit(X_train, Y_train)
    
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators= 10, criterion= 'entropy', random_state=0)
    forest.fit(X_train, Y_train)
    
    #printing accurecy of each model on the trainning dataset
    
    print("the accurecy of Logestic regression: ", log.score(X_train, Y_train))
    print("the accurecy of Decision tree: ", tree.score(X_train, Y_train))
    print("the accurecy of Random forest: ", forest.score(X_train, Y_train))
    
    return log , tree , forest 
```

```python
model = model(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, model[0].predict(X_test))

tp = cm[0][0]  #true positive
tn = cm[1][1]  #true negetive 
fp = cm[0][1]  #false positve
fn = cm[1][0]  #false negetive 

print(cm)
print('Accurecy', (tp+tn)/(tp+tn+fp+fn))

```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print('Model:',i)
    print(classification_report(Y_test, model[i].predict(X_test)))
    print()
```
